```python
import os
import arxiv
from pydantic import BaseModel, Field
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from dotenv import load_dotenv

# Load your Groq API key from .env
load_dotenv()

# 1. Define the desired JSON structure using Pydantic
class ArxivSearchParams(BaseModel):
    query: str = Field(
        description="The exact search keywords extracted from the user query. "
                    "If it's a specific phrase, enclose it in double quotes, e.g. 'all:\"machine learning\"'."
    )
    sort_by: str = Field(
        description="Must be exactly 'SubmittedDate' if the user explicitly asks for the 'latest', 'newest', "
                    "or 'recent' papers. Otherwise, default strictly to 'Relevance'."
    )
    max_results: int = Field(
        description="The number of papers the user wants. If not specified in the query, default to 5.",
        default=5
    )

def generate_arxiv_args(user_query: str) -> dict:
    """
    Uses Groq to parse a natural language query into arxiv search arguments.
    """
    # 2. Initialize the Groq model
    llm = ChatGroq(model_name="llama-3.1-8b-instant", temperature=0)
    
    # Bind the Pydantic model to force the LLM to output valid JSON matching our schema
    structured_llm = llm.with_structured_output(ArxivSearchParams)

    # 3. Create a prompt to instruct the LLM
    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are an expert research librarian. Your job is to convert natural language "
                   "requests into exact parameters for the arxiv python API.\n"
                   "- Extract the core technical keywords.\n"
                   "- If the user asks for the latest/newest papers, set sort_by to 'SubmittedDate'. "
                   "If they don't care about date or just want the best papers, set it to 'Relevance'."),
        ("human", "{query}")
    ])

    chain = prompt | structured_llm
    
    # 4. Run the chain
    print(f"Analyzing query: '{user_query}'...")
    result: ArxivSearchParams = chain.invoke({"query": user_query})
    
    # 5. Map the string result to actual arxiv enum objects
    if result.sort_by == "SubmittedDate":
        sort_criterion = arxiv.SortCriterion.SubmittedDate
    else:
        sort_criterion = arxiv.SortCriterion.Relevance

    # Return a dictionary that can be unpacked into arxiv.Search(**args)
    return {
        "query": result.query,
        "max_results": result.max_results,
        "sort_by": sort_criterion,
        "sort_order": arxiv.SortOrder.Descending
    }

# ==========================================
# TEST CASES
# ==========================================

# Test 1: User just wants the best papers (Defaults to Relevance)
params_1 = generate_arxiv_args("Find me 3 good papers on quantum error correction.")
print("JSON Params 1:", params_1)
print("-" * 40)

# Test 2: User explicitly wants the latest papers (Uses SubmittedDate)
params_2 = generate_arxiv_args("What are the absolute newest papers about spatial transcriptomics?")
print("JSON Params 2:", params_2)

# ==========================================
# HOW TO USE IT WITH ARXIV
# ==========================================
# 
# search = arxiv.Search(**params_2)
# client = arxiv.Client()
# results = list(client.results(search))
# for paper in results:
#     print(paper.title)
```

```python
import os
import arxiv
from pydantic import BaseModel, Field
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from dotenv import load_dotenv

load_dotenv()

# 1. Ask the LLM only for the query and a Yes/No preference
class ArxivQuery(BaseModel):
    query: str = Field(
        description="The exact search keywords extracted from the user query. Enclose specific phrases in double quotes (e.g. 'all:\"Attention is All you need\"' or '\"Attention is All you need\"')."
    )
    newest_preference: str = Field(
        description="If the user explicitly asks for the latest or newest papers, answer 'Yes', otherwise answer 'No'."
    )

def search_arxiv_smart(user_query: str):
    llm = ChatGroq(model_name="llama-3.1-8b-instant", temperature=0)
    structured_llm = llm.with_structured_output(ArxivQuery)
    
    prompt = ChatPromptTemplate.from_messages([
                ("system", "You are a research assistant. Extract the core search string from the user's request. "
                   "CRITICAL: You must ALWAYS enclose the extracted search topic in double quotes to force an exact phrase match (e.g., 'all:\"machine learning\"'). "
                   "Also determine if the user explicitly prefers the newest/latest papers ('Yes' or 'No')."),
        ("human", "{query}")
    ])
    
    # 2. Invoke the LLM to get structured JSON output
    chain = prompt | structured_llm
    result: ArxivQuery = chain.invoke({"query": user_query})
    
    # 3. Determine sorting logic based on the LLM's Yes/No answer
    if result.newest_preference.lower() == "yes":
        sort_criterion = arxiv.SortCriterion.SubmittedDate
    else:
        sort_criterion = arxiv.SortCriterion.Relevance
        
    # 4. Search configuration with fixed max_results=5
    search = arxiv.Search(
        query=result.query,
        max_results=5,
        # sort_by=sort_criterion,
        # sort_order=arxiv.SortOrder.Descending
    )
    
    print(f"User Query: '{user_query}'")
    print(f"LLM Extracted Keywords: {result.query}")
    print(f"Newest Preference: {result.newest_preference} (Using {sort_criterion})")
    print("=" * 50)
    
    # 5. Execute the search and print titles
    client = arxiv.Client()
    results = list(client.results(search))
    
    if not results:
        print("No papers found.")
        return
        
    for i, paper in enumerate(results, 1):
        print(f"[{i}] {paper.title}")
    print("\n")


# --- Testing both queries ---
search_arxiv_smart('what is quantum computiong')
# search_arxiv_smart('Find the newest papers on "Attention is All you need"')
```

```python
import os
import requests
import uuid
import arxiv

from pydantic import BaseModel, Field
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter

from qdrant_client import QdrantClient
from qdrant_client import models
from fastembed import LateInteractionTextEmbedding, SparseTextEmbedding

# 1. Define the desired JSON structure for the LLM selection
class PaperSelection(BaseModel):
    selected_index: int = Field(description="The index (1 to N) of the most relevant paper from the provided list.")
    reasoning: str = Field(description="A brief explanation of why this paper was chosen as the most relevant.")

def process_best_arxiv_paper(topic: str, max_results: int = 5):
    # --- Step 1: Search arXiv ---
    exact_query = f'all:"{topic}"'
    client = arxiv.Client()
    search = arxiv.Search(
        query=exact_query,
        max_results=max_results,
        sort_order=arxiv.SortOrder.Descending
    )
    
    results = list(client.results(search))
    if not results:
        print("No papers found matching that exact topic.")
        return
        
    print(f"Found {len(results)} papers. Asking LLM to pick the most relevant one...\n" + "="*50)
    
    # Prepare the list of titles to send to the LLM
    titles_context = ""
    for i, paper in enumerate(results, 1):
        titles_context += f"[{i}] Title: {paper.title}\n"
        
    # --- Step 2: Ask LLM to pick the most relevant paper ---
    llm = ChatGroq(model_name="llama-3.1-8b-instant", temperature=0)
    structured_llm = llm.with_structured_output(PaperSelection)
    
    prompt = ChatPromptTemplate.from_messages([
        ("system", "You are an expert AI research assistant. Given a specific topic and a list of paper titles retrieved from a search, select the single most relevant paper. Output the integer index of your choice and a brief reasoning."),
        ("human", "Topic: {topic}\n\nPapers:\n{titles}")
    ])
    
    chain = prompt | structured_llm
    
    # Invoke the LLM
    selection = chain.invoke({
        "topic": topic, 
        "titles": titles_context
    })
    
    # Get the selected paper (adjusting for 0-based indexing)
    best_paper = results[selection.selected_index - 1]
    
    print(f"LLM Selected: [{selection.selected_index}] {best_paper.title}")
    print(f"Reasoning: {selection.reasoning}")
    print(f"PDF URL: {best_paper.pdf_url}")
    print("="*50)
    
    # --- Step 3: Download PDF ---
    pdf_path = "best_paper.pdf"
    print("\nDownloading PDF...")
    response = requests.get(best_paper.pdf_url, headers={"User-Agent": "Mozilla/5.0"})
    
    # Basic Validation
    if response.status_code == 200 and response.content[:5] == b"%PDF-":
        with open(pdf_path, "wb") as f:
            f.write(response.content)
    else:
        print("Failed to download a valid PDF.")
        return
        
    # --- Step 4: Extract and Chunk Text ---
    print("\nExtracting and chunking text...")
    loader = PyPDFLoader(pdf_path)
    docs = loader.load()
    
    # Chunk the document (adjust sizes based on your experiment_pipeline1 settings)
    splitter = RecursiveCharacterTextSplitter(
        chunk_size=1000,
        chunk_overlap=200
    )
    chunks = splitter.split_documents(docs)
    text_chunks = [chunk.page_content for chunk in chunks]
    print(f"Created {len(text_chunks)} chunks from the PDF.")
    
    # --- Step 5: Embeddings and Qdrant ---
    print("\nInitializing embedding models...")
    colbert_model = LateInteractionTextEmbedding(model_name="colbert-ir/colbertv2.0")
    bm25_model = SparseTextEmbedding(model_name="Qdrant/bm25")
    
    print("Generating embeddings (this may take a moment)...")
    colbert_embeddings = list(colbert_model.embed(text_chunks))
    bm25_embeddings = list(bm25_model.embed(text_chunks))
    
    print("Connecting to Qdrant...")
    # Change to your actual Qdrant URL if you are running a local/cloud instance
    qdrant = QdrantClient(":memory:") 
    
    collection_name = "arxiv_papers"
    
    # Create the collection if it doesn't exist
    if not qdrant.collection_exists(collection_name):
        qdrant.create_collection(
            collection_name=collection_name,
            vectors_config={
                "colbert": models.VectorParams(
                    size=128,
                    distance=models.Distance.COSINE,
                    multivector_config=models.MultiVectorConfig(
                        comparator=models.MultiVectorComparator.MAX_SIM
                    )
                )
            },
            sparse_vectors_config={
                "bm25": models.SparseVectorParams(modifier=models.Modifier.IDF)
            }
        )
    
    # Build points
    points = []
    paper_id = best_paper.entry_id.split('/')[-1]
    
    for idx, text in enumerate(text_chunks):
        point_id = str(uuid.uuid4())
        points.append(
            models.PointStruct(
                id=point_id,
                payload={
                    "text": text,
                    "paper_title": best_paper.title,
                    "paper_id": paper_id
                },
                vector={
                    "colbert": colbert_embeddings[idx].tolist() if hasattr(colbert_embeddings[idx], 'tolist') else colbert_embeddings[idx],
                    "bm25": models.SparseVector(
                        indices=bm25_embeddings[idx].indices.tolist() if hasattr(bm25_embeddings[idx].indices, 'tolist') else bm25_embeddings[idx].indices,
                        values=bm25_embeddings[idx].values.tolist() if hasattr(bm25_embeddings[idx].values, 'tolist') else bm25_embeddings[idx].values,
                    )
                }
            )
        )
        
    print("Uploading points to Qdrant...")
    qdrant.upload_points(
        collection_name=collection_name,
        points=points
    )
    
    print("\n✅ Done! The paper has been successfully processed, embedded, and stored.")

# --- Test the pipeline ---
# process_best_arxiv_paper("Attention is All you need", max_results=5)
```
