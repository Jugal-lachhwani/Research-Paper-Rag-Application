```python
import os
import time
import requests
import arxiv

# Create arXiv client
client = arxiv.Client(
    page_size=10,
    delay_seconds=3,
    num_retries=3
)

# Search
search = arxiv.Search(
    query='ti:"Attention Is All You Need"',
    max_results=1
)

paper = next(client.results(search))

print("Title:", paper.title)
print("PDF URL:", paper.pdf_url)

# Download
response = requests.get(
    paper.pdf_url,
    headers={
        "User-Agent": "Mozilla/5.0"
    },
    timeout=60
)

print("Status Code:", response.status_code)
print("Content-Type:", response.headers.get("Content-Type"))

# Verify PDF
if response.content[:5] != b"%PDF-":
    raise Exception("Downloaded file is NOT a PDF!")

filename = "attention.pdf"

with open(filename, "wb") as f:
    f.write(response.content)

print(f"Saved as {os.path.abspath(filename)}")

time.sleep(3)
```

```python
import arxiv

client = arxiv.Client()

search = arxiv.Search(
    query='all:"quantum computing"',
    max_results=10,
    sort_by=arxiv.SortCriterion.SubmittedDate,
    sort_order=arxiv.SortOrder.Descending,
)

papers = list(client.results(search))

for paper in papers:
    print(paper.title)
    print(paper.published)
    print(paper.pdf_url)
    print()
```
